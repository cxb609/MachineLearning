```python
import pandas as pd
import numpy as np
import requests
r = requests.get('''https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/regression/housing_scale''')

def loss_function(x,y,w):
    x_shape = x.shape
    m = x_shape[0]
    xw = x@w
    yy = []
    for i in y:
        yy.append(i)
    sum = 0
    for i in range(0,x_shape[0]):
        sum = sum + (yy[i]-xw[i])*(yy[i]-xw[i])
            
    regular = 0
    for j in range(0,x_shape[1]):
        regular = regular + w[j]*w[j]
    sum = sum/(2*m) + 1/2*regular
    
    loss = sum
    return loss

def grad(x,y,w):
    temp = -1*(x.T.dot(y))
    l = []
    for i in temp:
        ll = []
        ll.append(i)
        l.append(ll)
    return l+x.T.dot(x).dot(w)

from sklearn.datasets import load_svmlight_file
from io import BytesIO
X, Y = load_svmlight_file(f=BytesIO(r.content), n_features=13)
X = X.toarray()
Y = Y.reshape((-1, 1))
x_shape = X.shape
x0 = np.ones((x_shape[0],1))
X = np.column_stack((X,x0))

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)

W = np.random.rand(x_shape[1]+1,1)
train_shape = X_train.shape

losses_train = []
losses_val = []

import random
for count in range(0,400):
    # choose a training sample randomly
    i = random.randint(0,train_shape[0]-1)
    x_sample = X_train[i:i+1,:]
    yy = []
    count = 0
    for ii in Y_train:
        yy.append(ii)
    y_sample = yy[i]
    
    d = -1*grad(x_sample,y_sample,W)
    W = W + 0.001*d
    
    loss_train = loss_function(X_train,Y_train,W)
    losses_train.append(loss_train)
    
    loss_val = loss_function(X_test,Y_test,W)
    losses_val.append(loss_val)

import matplotlib.pyplot as plt

plt.figure(figsize=(18, 6))
plt.plot(losses_train, "-", color="r", label="train loss")
plt.plot(losses_val, "-", color="b", label="validation loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.title("The graph of absolute diff value varing with the number of iterations")
```
